# Section 4: Handling Missing Values

```python
# using the below doesn't count null values
drinks.continent.value_counts()
```

```python
# Use dropna=False command for the value_counts to count Null values
drinks.continent.value_counts(dropna=False)
```

```python
# Look at the sum of Null values for column 'continent' using isnull() command
drinks.continent.isnull().sum()
```

```python
# Look at the sum of Null values for column 'continent' using notnull() command
drinks.continent.notnull()
```

```python
# Filter out Null values using mask
mask = drinks.continent.notnull()
drinks[mask]
```

```python
# explore rows in the dataframe
drinks.sum(axis=0)
```

```python
# explore columns in the dataframe - adding columns for each row
drinks.sum(axis=1)
```

```python
drinks['wine_litres'] = drinks[['wine', 'litres']].sum(axis=1)
drinks.wine_litres
```

```python
drinks.columns
```

```python
drinks.info()
```

```python
drinks.isnull().sum().plot(kind='bar')
```

```python
# drop any row that has Null in every column
drinks.dropna()
```

```python
# drop
drinks.dropna(how='all')
```

```python
drinks.dropna(inplace = True)
```

```python
drinks
```

```python
drinks.reset_index()
```

```python
drinks.continent.isnull()
```

```python
#replace Null with NA

drinks.continent.fillna(value='NA', inplace=True)
```
