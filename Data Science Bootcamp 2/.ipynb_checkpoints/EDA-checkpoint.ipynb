# Counting the number of occurences for each value

```python
users
```

```python
# value_counts is a command for categorical values
users.gender.value_counts()
```

```python
users.gender.value_counts().plot(kind='bar')
```

```python
# Value_counts used on numerical values will show frequency for each integer
users.age.value_counts()
```

```python
users.age.value_counts().sort_index()
```

```python
users.age.value_counts().sort_index().plot(kind='bar')
```

```python
users.age.value_counts().sort_index().plot(kind='bar',figsize=(12,8))
plt.title('Distribution of ages')
plt.xlabel('Age')
plt.ylabel('Number of users per age')
```

# Logical Filtering with multiple conditions

```python
drinks
```

```python
# What are the European countries that drink a lot of beer

mask = (drinks.beer_servings > 25) & (drinks.continent == 'EU')
```

```python
mask
```

```python
drinks[mask]
```

```python
# What are the countries that drink a lot of beer OR in Europe

mask = (drinks.beer_servings > 25) | (drinks.continent == 'EU')
```

```python
drinks[mask]
```

```python
# What are the countries that drink a lot of beer OR in Europe

mask = ((drinks.beer_servings > 25) & (drinks.continent == 'EU')) | (drinks.wine_servings > 200)
```

```python
drinks[mask]
```
