# Split - Apply - Combine

```python
# Split method - group data by a specific Value
# Apply method - for each group apply 
# Combine method - combine the Split and Apply methods
```

```python
drinks = pd.read_csv('Data/drinks.csv')
```

```python
drinks
```

```python
# get average of beer_servings for each continent 

drinks.groupby('continent').beer_servings.mean()
```

```python
# for each continent calculate the average of every column
drinks.groupby('continent').mean()
```

```python
# for beer_servings get a descriptive stats for each continent
drinks.groupby('continent').beer_servings.describe()
```

```python
# for beer_servings get an agreegate descriptive stats for each continent -  Agg allows us to aggregate results of a series function
drinks.groupby('continent').beer_servings.agg(['mean','count','min','max'])
```

```python
# Sort dataframe by Mean
drinks.groupby('continent').beer_servings.agg(['mean','count','min','max']).sort_values('mean')
```

```python
# Sort dataframe by Mean in the descending order
drinks.groupby('continent').beer_servings.agg(['mean','count','min','max']).sort_values('mean', ascending = False)
```

```python
# counting number of rows for every 'continent' column
drinks.groupby('continent').continent.count()
```

```python
# the command above is the same as using value_counts()
drinks.continent.value_counts()
```
