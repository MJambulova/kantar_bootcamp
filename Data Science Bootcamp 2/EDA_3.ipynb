# Joining methods

```python
# Load the Movies dataset

movie_cols = ['movie_id', 'title']
movies_filename = 'data/movies.tbl'

movies = pd.read_table(
    movies_filename,
    sep='|',
    header=None,
    names=movie_cols,
    usecols=[0, 1],
    encoding='latin-1')

movies
```

```python
# Load the Ratings dataset

rating_cols = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings_filename = 'data/movie_ratings.tsv'

ratings = pd.read_table(ratings_filename, sep='\t', header=None, names=rating_cols)
ratings.head()

ratings
```

```python
# merge movies table and ratings table
movie_ratings = pd.merge(movies, ratings)
```

```python
movie_ratings
```

```python
# Return the average rating for every movie title and sort it Z to A with the top 10
movie_ratings.groupby('title').rating.mean().sort_values(ascending=False).head(10)
```

# Apply

```python
drink_cols = ['country', 'beer', 'spirit', 'wine', 'liters', 'continent'] 

drinks = pd.read_csv('data/drinks.csv', header=0, names=drink_cols)
```

```python
drinks
```

```python
# Classify countries by the liters as Low and Large - first look for the median liter 

median = drinks.liters.median()
median
```

```python
# Apply function works as a for loop function by looking through every row and applying function to the specific row
# Apply function is usually used with the Lambda function 
```

```python
# Create a new column 'volume' by using apply function which will look into the liters column and take every row and multiply it by 2

drinks['volume'] = drinks.liters.apply(lambda x: x*2 )
drinks
```

```python
# add a column 'volume' where every row is classed as Large or low based on the liters median

drinks['volume'] = drinks.liters.apply(lambda x: 'large' if x > median else 'low')
drinks
```

```python
# using Lambda and apply for multiple conditions

drinks['volume'] = drinks.liters.apply(lambda x: 'very large' if x > 8 else 'large' if x > median else 'low')
drinks
```

```python
drinks.dtypes
```

```python
# change liters datatype to string. Object datatype - is any other datatype that is not recognised by Pandas
drinks['liters'] = drinks.liters.apply(lambda x: int(x))
```

```python
# Create a function that takes an input and rounds it to the nearest whole number using round(x,0)
# Use ''' ''' for comments that are on multiple lines

def round_whole(x):
    '''
    INPUT: this function takes input a single number
    OUTPUT: the output is a rounded number
    '''    
    rounder = round(x,0)
    return rounder
```

```python
drinks['liters'] = drinks.liters.apply(lambda x: round_whole(x))
drinks
```

```python
# change the liters datatype to integer or round 

drinks['litres'] = drinks.litres.apply(round)
drinks['litres'] = drinks.litres.apply(int)
```
